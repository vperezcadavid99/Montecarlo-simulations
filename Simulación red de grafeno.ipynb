# Introducción

A pesar de que el grafeno en su forma pura es intrínsecamente no magnético debido a su estructura cristalina de carbono sp², se ha demostrado que puede exhibir propiedades magnéticas cuando se introducen defectos, impurezas o se modifican sus condiciones ambientales. Estos defectos pueden incluir la presencia de átomos de impurezas o la deformación de la red cristalina, lo que puede inducir momentos magnéticos locales y generar interesantes fenómenos magnéticos, como el ferromagnetismo o paramagnetismo.

<p align="justify">

# Marco Teórico
## Paramagnetismo de Ising

<p align="justify">
El modelo propuesto por Ernst Ising en 1920, implica un sistema compuesto por N partículas, cada una con un momento magnético intrínseco distinto de cero, dispuestas en los nodos de una red regular. Aunque inicialmente estos momentos podrían alinearse en cualquier dirección, el modelo simplifica esta posibilidad considerando que solo pueden orientarse a lo largo de un eje cartesiano específico, como el eje z, con una polarización positiva o negativa, es decir, hacia "arriba" o "abajo", respectivamente. Este atributo de la orientación de las partículas es denominado "espín", en una clara analogía con el concepto cuántico, y puede tomar valores de $\sigma \pm 1/2$.

<p align="justify">
Los espines pueden estar distribuidos de manera aleatoria, lo que resulta en una magnetización nula (paramagnetismo) en ausencia de campo, o pueden ordenarse de tal manera que se genere una magnetización neta (ferromagnetismo). La descripción de esta configuración de espines en términos del Hamiltoniano de la i-ésima partícula se expresa como:

Ec(1) $$H_{i} = -J\sigma _{i}\sum_{<j>}^{}\sigma _{j}-\sigma _{i}H$$

<p align="justify">
Donde el primer término da cuenta de la interacción del $i$-ésimo espín con sus primeros vecinos, y el segundo término indica la interacción de este mismo espín con un campo magnético $H$ externo aplicado al sistema, conocido también como término Zeeman. En la ecuación 1, se destaca la influencia de la integral de intercambio $J$, que refleja el grado de superposición de las funciones de onda asociadas a cada partícula, debido a las interacciones entre partículas vecinas. La variación de este término permite clasificar los sistemas magnéticos de la siguiente manera:


<p align="justify">

* Cuando J = 0, indica que no hay interacción entre los momentos magnéticos del
  sistema, lo que resulta en una magnetización nula en ausencia de un campo magnético externo aplicado. Estos materiales se denominan paramagnéticos.

<p align="justify">

*   Por otro lado, si J > 0, implica que existe una interacción que alinea los
    momentos magnéticos en una misma dirección respecto a un campo externo aplicado, lo que caracteriza a los materiales ferromagnéticos.

<p align="justify">

*   Finalmente, si J < 0, indica una interacción donde los momentos magnéticos
    permanecen orientados antiparalelamente tanto en ausencia como en presencia de un campo externo aplicado. Este tipo de comportamiento se observa en materiales antiferromagnéticos.


## Ferromagnetismo

<p align="justify">
En los materiales ferromagnéticos, los dipolos magnéticos atómicos exhiben una alineación parcial, lo que genera una magnetización intrínseca cuando no hay un campo magnético externo $H$. Sin embargo, al exponer estos materiales a un campo magnético externo, los momentos magnéticos tienden a alinearse en la dirección del campo aplicado, lo que conduce a un aumento sustancial de la magnetización [1].

<p align="justify">
Debido al fuerte acoplamiento ente el campo y los momentos magnéticos de este tipo de sistemas, el efecto de la aparición de la magnetización intrínseca en ausencia de campo es conocido como histéresis. Esto es evidente en las curvas de magnetización en función del campo aplicado.

Se pueden distinguir tres puntos claves


*   Campo coercitivo ($H_c$): valor de campo magnético necesario para anular la
    imanación.

*   Magnetización de saturación: se consigue al aplicar un campo magnético
    suficiente mente grande para lograr la máxima alineación de los momentos magnéticos, generando un valor de magnetización máximo.

*   Magnetización remanente: magnetización que permanece en un sistema incluso  cuando no hay campo magnético aplicado, después de haber sido expuesto a un campo magnético lo suficientemente intenso como para saturar su respuesta magnética.


## Algoritmo de Metropolis

<p align="justify">
El procedimiento se basa en las llamadas cadenas de Markov, estas representan una secuencia de estados en la que, en cada paso, el sistema puede cambiar de su estado actual a uno nuevo con cierta probabilidad de transición. Esta probabilidad de transición solo depende del estado actual y no se ve afectada por la historia del proceso. Si estas probabilidades se eligen de manera adecuada, cada estado eventualmente aparecerá con la frecuencia deseada después de un período de equilibrio, según el factor de Boltzmann, ya que buscamos un sistema en equilibrio termodinámico. Es importante destacar que en este contexto, "tiempo" se refiere a los pasos de la simulación y no al tiempo real. describe como sigue.

<p align="justify">
En este orden de ideas, el algoritmo se describe como sigue. En primer lugar, se selecciona una configuración inicial del sistema, lo que podría implicar asignar un valor aleatorio a cada espín, reflejando así una configuración con temperatura finita. Luego, en cada iteración se elige al azar un espín en la red y se intenta cambiar su orientación, lo que altera la configuración actual σ con energía $E_1$ a una nueva configuración τ con energía $E_2$. A continuación, se evalúa si se acepta o no la perturbación. Para esto, se calcula la diferencia de energía ∆E entre las configuraciones. Si $∆E ≤ 0$, la perturbación se acepta automáticamente. Sin embargo, si $∆E > 0$, se acepta el cambio con una probabilidad de $e^{(-β∆E)}$, lo que corresponde a:

Ec(2)
$$\frac{W(E_1 \rightarrow E_2 )}{W(E_2\rightarrow E_1 )} = \frac{e^{-\beta E_2 }}{e^{-\beta E_1 }} = e^{-\beta \Delta E}$$


<p align="justify">
en caso contrario, se rechaza el cambio, manteniendo al sistema en su configuración actual σ. Este proceso se repite un gran número de veces para asegurar que las configuraciones generadas reflejen adecuadamente la probabilidad deseada, correspondiente al equilibrio termodinámico [2].

# Resultados

<p align="justify">
En primer lugar se construye una red hexagonal de tamaño $L_i \times L_j$ con condicione de frontera periódicas, la cual obedece la teoría de Ising Ec1, en la cual se ubican de forma aleatoria los $n$ spines "up" o "down" teniendo en cuenta el parámetro de ocupación q. En este caso se utiliza una red $10 \times 10$.

```python
#Importar librerias necesarias
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

#Definir tamaño de la red en Número de Hexagonos verticales x Número de Hexagonos horizontales
L_i = 10
L_j = 10

#Ubicación de nodo ij-ésimo basado en el tamaño definido
j = 2*(L_j + 1) + 1
i = 2*(L_i + 1)

#Inicialización de arreglo de ceros
red = np.zeros((i,j))

#Ciclos anidados utilizados para llenar todos los nodos posibles con el número 2, construyendo
#así la red inicial
m = 0
for k in range(0,i,2):
  for l in range(0,j,2):
    if m == 0:
      red[k][l] = 2
      if l < j-1: red[k+1][l+1] = 2
    else:
      red[k+1][l] = 2
      if l < j-1: red[k][l+1] = 2
  m=1 if m==0 else 0
```

```python
#Función para aplicar condiciones de frontera periodicas, teniendo en cuenta paridad en el número
#de Hexagonos, la función regresa jp1 = i+1 y jm1 = j-1 para ser utilizado en la inspección de la red
def pbc(i,j):
  J = 2*(L_j + 1) + 1
  I = 2*(L_i + 1)

  if L_i%2 == 0:
    im1 = i-1 if i-1 >=0 else I-2
    ip1 = i+1 if i+1 < I else 1
  else:
    im1 = i-1 if i-1 >= 0 else I-1
    ip1 = i+1 if i+1 < I else 0

  jm1 = j-1 if j-1 >= 0 else J-2
  jp1 = j+1 if j+1 < J else 1

  return ip1,im1,jp1,jm1

#Construcción de la red real, dependiendo de un porcentaje de ocupación
def build_system(red,q):
  #Número de nodos
  N = np.sum(red)/2
  #Número de nodos ocupados
  n = q*N
  red_spin = red.copy()

  r = 0
  while r < n:
    #Ubicación ij-ésica del spín de forma aleatoria
    k = np.random.randint(0,i)
    l = np.random.randint(0,j)
    if red_spin[k][l] == 2:
      #Definición de spin up o down de forma aleatoria
      value = np.random.randint(0,2)
      if value == 0: value = -1
      red_spin[k][l] = value
      r += 1
    else:
      continue
  #Reemplazo de nodos vacios por ceros para simplificar cáculos de energía
  for h in range(i):
    for s in range(j):
      if red_spin[h][s] == 2: red_spin[h][s] = 0

  return red_spin, n

#Cálculo de la energía por spin
def energy_spin(red,i,j,H,J_exc):
  #Si el nodo está vacío la energía es cero
  if red[i][j] == 0:
    energy = 0
  else:
    #Condiciones de frontera periodicas
    ip1,im1,jp1,jm1 = pbc(i,j)

    #Cálculo de la energía dependiendo de la configuración de los primeros vecinos
    if i%2 == 0:
      energy = -J_exc * red[i][j] * (red[ip1][jp1] + red[ip1][jm1]+ red[im1][j]) - red[i][j] * H
    else:
      energy = -J_exc * red[i][j] * (red[im1][jp1] + red[ip1][j] + red[im1][jm1]) - red[i][j] * H

  return energy
```

<p align="justify">
Adicionalmente se define una función para el cálculo de la energía por sitio magnético obedeciendo la Ec1. Teniendo en cuenta la ubicación de los primeros vecinos y la interacción con un campo magnético externo $H$.

<p align="justify">
Finalmente, se define el algoritmo de Metropolis-Hastings como método de montecarlo, en la que se evoluciona el sistema teniendo en cuenta las perturbación introducidas a partir de las condiciones dadas por la Ec2.

```python
#Definición de algoritmo de Metropolis-Hastings
def montecarlo(STEPS,red_spin,T,H,J_exc):
  #Arreglos para la magnetización y energía en cada paso, y sus cuadrados
  Mag = []
  Em = []
  Mag_cuad = []
  Em_cuad = []

  #Obtención de únicamente los índices de la red que contengan spin
  new_only_spin = red_spin.copy()
  indices = np.where(np.logical_or(red_spin == -1, red_spin == 1))

  for _ in range(STEPS):

    #Cálculo de la energía total de la red para cada paso, y su cuadrado
    E = 0
    for l in range(len(indices[0])):
      E += energy_spin(new_only_spin, indices[0][l], indices[1][l],H, J_exc)
    Em.append(E/2)
    Em_cuad.append((E/2)**2)

    for k in range(len(indices[0])):
      #Cálculo del delta de energía tras realizar una perturbación en el spin ij-ésimo
      Delta_E = -2*energy_spin(new_only_spin, indices[0][k], indices[1][k], H, J_exc)

      #Condiciones definidas en el algoritmo para aceptar la perturbación como minimizadora de energía
      if Delta_E <= 0:
        new_only_spin[indices[0][k]][indices[1][k]] *= -1
      else:
        r = np.random.uniform()
        W = np.exp(-(1/T) * Delta_E)
        if r <= W:
          #Reemplazo de nuevo valor de spin en la red
          new_only_spin[indices[0][k]][indices[1][k]] *= -1

    #Cálculo de la magnetización total de la red para cada paso, y su cuadrado
    Mag.append(np.sum(new_only_spin))
    Mag_cuad.append(np.sum(new_only_spin)**2)

  return Em, Em_cuad, Mag, Mag_cuad, new_only_spin
```

<p align="justify">
La presencia de cambios abruptos en las curvas de histéresis de magnetización, con líneas verticales y horizontales, puede ser indicativa de la presencia de anisotropía magnética inducida en el sistema. Este fenómeno puede surgir debido a la inhomogeneidad en la concentración de ocupación de espines en el material.

<p align="justify">
Cuando se aplica un campo magnético externo, estas regiones con diferentes orientaciones magnéticas pueden responder de manera distinta, lo que puede resultar en cambios abruptos en la magnetización del material a lo largo de la curva de histéresis. [3]

<p align="justify">
Con el fin de estudiar la dependencia de la temperatura de transición con el porcentaje de ocupación q. Es importnate primero estudiar la distribución de espines en la red a medida que aumenta la temperatura. A continuación se presenta la evolución del sistema para $q = 1.0, 0.8, 0.5, 0.1$, establaciendo un campo de $H = 0$ y una integral de interacción $J = 5$

```python
#Construcción del sistema con únicamente spines up
def build_system_up(red,q):
  #Número de nodos
  N = np.sum(red)/2
  #Número de nodos ocupados
  n = q*N
  red_spin = red.copy()

  r = 0
  while r < n:
    #Ubicación ij-ésica del spín de forma aleatoria
    k = np.random.randint(0,i)
    l = np.random.randint(0,j)
    if red_spin[k][l] == 2:
      #Ubicación de spin up
      red_spin[k][l] = 1.0
      r += 1
    else:
      continue
  #Reemplazo de nodos vacios por ceros para simplificar cáculos de energía
  for h in range(i):
    for s in range(j):
      if red_spin[h][s] == 2: red_spin[h][s] = 0

  return red_spin, n
```

```python
#Redefinición de la red para realizar análisis sobre tamaños más grandes
L_i = 16
L_j = 16

#Ubicación de nodo ij-ésimo basado en el tamaño definido
j = 2*(L_j + 1) + 1
i = 2*(L_i + 1)

#Inicialización de arreglo de ceros
red_snapshots = np.zeros((i,j))

#Ciclos anidados utilizados para llenar todos los nodos posibles con el número 2, construyendo
#así la red inicial
m = 0
for k in range(0,i,2):
  for l in range(0,j,2):
    if m == 0:
      red_snapshots[k][l] = 2
      if l < j-1: red_snapshots[k+1][l+1] = 2
    else:
      red_snapshots[k+1][l] = 2
      if l < j-1: red_snapshots[k][l+1] = 2
  m=1 if m==0 else 0
```

```python
#Definición de red para análisis de magnetización en función de la temperatura
red_spin_0p5, n_0p5 = build_system_up(red,1.0)
#Arreglo de temperaturas a evaluar
T = np.linspace(0.5,30,100)

#Pasos de montecarlo
STEPS = 3000

#Listas para guardar magnetizaciónes y energías medias, así como sus cuadrados
m_0p5_5 = []
m_cuad_0p5_5 = []
E_0p5_5 = []
E_cuad_0p5_5 = []

#Ciclo para calcular la evolución temporal del sistema con montecarlo
for t in tqdm(T):
  Em_0p5_5, Em_cuad_0p5_5, Mag_0p5_5, Mag_cuad_0p5_5, spin_perturbation_0p5_5 = montecarlo(STEPS,red_spin_0p5,t,0,5)

  #Cálculo de valor esperado de la magnetización y su cuadrado para cada paso
  esp_m_0p5_5 = (1/(STEPS - 1000)) * (np.sum(Mag_0p5_5[1000+1:])) * (1/n_0p5)
  esp_m_cuad_0p5_5 = (1/(STEPS - 1000)) * (np.sum(Mag_cuad_0p5_5[1000+1:])) * (1/n_0p5)**2
  m_0p5_5.append(esp_m_0p5_5)
  m_cuad_0p5_5.append(esp_m_cuad_0p5_5)

  #Cálculo de valor esperado de la energía y su cuadrado para cada paso
  esp_E_0p5_5 = (1/(STEPS - 1000)) * (np.sum(Em_0p5_5[1000+1:]))
  esp_E_cuad_0p5_5 = (1/(STEPS - 1000)) * (np.sum(Em_cuad_0p5_5[1000+1:]))
  E_0p5_5.append(esp_E_0p5_5)
  E_cuad_0p5_5.append(esp_E_cuad_0p5_5)

#Cálculo de varianzas para magnetización y energía
var_m_0p5_5 = np.array(m_cuad_0p5_5) - np.array(m_0p5_5)**2
var_E_0p5_5 = np.array(E_cuad_0p5_5) - np.array(E_0p5_5)**2
```
